**Xg boostin,Gradient boosting,ADaboostin**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Classification
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_classes=2,n_features=4,n_informative=2,n_redundant=0,random_state=1)
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
# import Xg,ada nd gradient boost
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
xg=XGBClassifier()
ada=AdaBoostClassifier()
gb=GradientBoostingClassifier()
```

```python
xg.fit(X_train,y_train)
```

```python
ada.fit(X_train,y_train)
```

```python
# prediction
y_pred=ada.predict(X_test)
```

```python
# performance of the model
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
xg.fit(X_train,y_train)
```

```python
y_pre=xg.predict(X_test)
```

```python
# performance
print(accuracy_score(y_test,y_pre))
print(confusion_matrix(y_test,y_pre))
print(classification_report(y_test,y_pre))
```

```python
gb.fit(X_train,y_train)
```

```python
y_pret=gb.predict(X_test)
```

```python
#performance
print(accuracy_score(y_test,y_pret))
print(confusion_matrix(y_test,y_pret))
print(classification_report(y_test,y_pret))
```

```python
# regression using xg,ada,gradient
from sklearn.datasets import make_regression
```

```python
L,m=make_regression(n_samples=1000,n_features=4,n_informative=2,n_targets=1,noise=0.1,random_state=42)
```

```python
# train and test split
from sklearn.model_selection import train_test_split
L_train,L_test,m_train,m_test=train_test_split(L,m,test_size=0.3,random_state=1)
```

```python
# train the model using xg,ada,gradient regressor
from xgboost import XGBRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# make objects
xg=XGBRegressor()
ada=AdaBoostRegressor()
gb=GradientBoostingRegressor()
```

```python
xg.fit(L_train,m_train)
```

```python
# predict
m_predt=xg.predict(L_test)
```

```python
# performnce of the model
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
score=r2_score(m_test,m_predt)
print(score)
```

```python
ada.fit(L_train,m_train)
```

```python
# predict
m_predi=ada.predict(L_test)
```

```python
# performance of the moel
score=r2_score(m_test,m_predi)
print(score)
```

```python
gb.fit(L_train,m_train)
```

```python
# prediction
m_pret=gb.predict(L_test)
```

```python
# performance
score=r2_score(m_test,m_pret)
print(score)
```
